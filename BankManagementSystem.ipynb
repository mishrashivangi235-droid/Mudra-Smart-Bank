```python
import json
import random
import string
from pathlib import Path

class Bank:
    # Changed to a relative path so it runs anywhere
    database = 'bank_data.json'
    data = []

    try:    
        if Path(database).exists():
            with open(database) as fs:
                content = fs.read().strip()
                data = json.loads(content) if content else []
        else:
            # Create an empty file if it doesn't exist
            with open(database, 'w') as fs:
                fs.write(json.dumps([]))
    except Exception as err:
        print(f"An exception occurred: {err}")
             
    @classmethod
    def __update(cls):
        with open(cls.database, 'w') as fs:
            fs.write(json.dumps(cls.data, indent=4))

    @classmethod
    def __accountgenerate(cls):
        alpha = random.choices(string.ascii_letters, k=3)
        num = random.choices(string.digits, k=3)
        spchar = random.choices("!@#$%^&*", k=1)
        id_list = alpha + num + spchar
        random.shuffle(id_list)
        return "".join(id_list)

    def Createaccount(self):
        try:
            name = input("Enter your name: ")
            age = int(input("Enter your age: "))
            email = input("Enter your Email: ")
            pin = int(input("Enter a 4-digit pin: "))
        except ValueError:
            print("Invalid input type! Age and PIN must be numbers.")
            return

        if age < 18 or len(str(pin)) != 4: 
            print("Sorry, you cannot create an account (Underage or invalid PIN).")
        else:
            info = {
               "name" : name,
               "age" : age,
               "email" : email,
               "pin" : pin,
               "AccountNo." : Bank.__accountgenerate(),
               "Balance" : 0
            }
            print("\nAccount has been created successfully!")
            for key, val in info.items():
                print(f"{key} : {val}")
            print("Please note down your account number.\n")
          
            Bank.data.append(info)
            Bank.__update()

    def _get_user(self):
        """Helper method to reduce code repetition across functions"""
        acc_num = input("Please enter your account no: ")
        try:
            pin = int(input("Please enter your pin: "))
        except ValueError:
            return None
        
        # Returns the user dictionary directly or None if not found
        for user in Bank.data:
            if user['AccountNo.'] == acc_num and user['pin'] == pin:
                return user
        return None

    def depositmoney(self):
        user = self._get_user()
        if not user:
            print("No matching data found.")
            return

        try:
            amount = int(input("How much money do you want to deposit? "))
        except ValueError:
            print("Invalid amount.")
            return

        if amount > 10000 or amount <= 0:
            print("Invalid amount. Deposit must be between 1 and 10,000.")
        else:
            user['Balance'] += amount
            Bank.__update()
            print(f"Amount deposited successfully. New Balance: {user['Balance']}")

    def withdrawmoney(self):
        user = self._get_user()
        if not user:
            print("No matching data found.")
            return

        try:
            amount = int(input("How much money do you want to withdraw? "))
        except ValueError:
            print("Invalid amount.")
            return

        if user['Balance'] < amount:
            print("Sorry, you don't have enough balance.")
        elif amount <= 0:
            print("Amount must be greater than 0.")
        else:
            user['Balance'] -= amount
            Bank.__update()
            print(f"Amount withdrawn successfully. Remaining Balance: {user['Balance']}")

    def showdetails(self):
        user = self._get_user()
        if not user:
            print("No matching data found.")
            return

        print("\nYour account information:")
        for key, val in user.items():
            print(f"{key} : {val}")

    def updatedetails(self):
        user = self._get_user()
        if not user:
            print("No matching data found.")
            return

        print("You cannot change age, account number, or balance.")
        print("Leave field empty and press Enter if no changes are needed.")

        new_name = input("Enter new name: ")
        new_email = input("Enter new email: ")
        new_pin = input("Enter new 4-digit pin: ")

        updated = False
        if new_name.strip():
            user['name'] = new_name
            updated = True
        if new_email.strip():
            user['email'] = new_email
            updated = True
        if new_pin.strip():
            if len(new_pin) == 4 and new_pin.isdigit():
                user['pin'] = int(new_pin)
                updated = True
            else:
                print("Invalid PIN format. PIN not updated.")

        if updated:
            Bank.__update()
            print("Details updated successfully!")
        else:
            print("No changes made.")

    def Delete(self):
        user = self._get_user()
        if not user:
            print("No matching data found.")
            return

        check = input("Press 'y' to confirm account deletion, or any other key to cancel: ")
        if check.lower() == 'y':
            Bank.data.remove(user)
            Bank.__update()
            print("Account deleted successfully.")
        else:
            print("Deletion canceled.")

                
# --- Main Execution Block ---
user = Bank()

print("--- Welcome to the Bank System ---")
print("1. Create Account")
print("2. Deposit Money")
print("3. Withdraw Money")
print("4. Show Account Details")
print("5. Update Account Details")
print("6. Delete Account")

try:
    check = int(input("Enter your response (1-6): "))
    print("-" * 30)
    
    if check == 1: user.Createaccount()
    elif check == 2: user.depositmoney()
    elif check == 3: user.withdrawmoney()
    elif check == 4: user.showdetails()
    elif check == 5: user.updatedetails()
    elif check == 6: user.Delete()
    else: print("Invalid option selected.")
except ValueError:
    print("Please enter a valid numeric choice.")
```
